```python
import pandas as pd
from statistics import mean
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import random
style.use('fivethirtyeight')
```

```python
x=np.array([1,2,3,4,5,6],dtype=np.float64)
y=np.array([5,4,6,5,6,7],dtype=np.float64)
plt.scatter(x,y)
plt.show()
```

```python
def best_fit_slope_intercept(x,y):
    m = ((mean(x)*mean(y))-mean(x*y))/((mean(x)*mean(x))-mean(x**2))
    b=mean(y)-m*mean(x)
    return m,b
m,b=best_fit_slope_intercept(x,y)
print(m,b)
```

```python
regression_line=[(m*i)+b for i in x]
```

```python
plt.scatter(x,y)
plt.plot(x,regression_line)
plt.show()
```

```python
def predict(x,m,b):
    y=[m*i+b for i in x]
    return y
predict_x=[8,9,10]
predict_y=predict(predict_x,m,b)
plt.scatter(x,y)
plt.scatter(predict_x,predict_y)
plt.plot(x,regression_line)
plt.show()
```

```python
def squared_error(y_original,y_line):
    return sum((y_line-y_original)**2)
def coef_of_determination(y_ori,y_line):
    y_mean_line = [mean(y_ori) for y in y_ori]
    squared_error_y_mean = squared_error(y_ori,y_mean_line)
    squared_error_regr = squared_error(y_ori,y_line)
    return (1-(squared_error_regr/squared_error_y_mean))
```

```python
r_squared = coef_of_determination(y,regression_line)
print(r_squared)
```
